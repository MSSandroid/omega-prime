```python
import numpy as np

import omega_prime
```

```python
r = omega_prime.Recording.from_file(
    "example_files/pedestrian.osi", xodr_path="example_files/fabriksgatan.xodr", parse_map=True
)
```

```python
locator = omega_prime.Locator.from_mapodr(r.map)
```

```python
xys = np.stack([r.moving_objects[0].x, r.moving_objects[0].y]).T
xys.shape
```

```python
sts = locator.xys2sts(xys)
sts
```

```python
sts_poly = locator.xys2sts(xys, poly=r.moving_objects[0].polygon)
sts_poly
```

```python
derived_xys = locator.sts2xys(sts)
derived_xys.shape
```

```python
derived_xys_poly = locator.sts2xys(sts_poly)
derived_xys_poly.shape
```

```python
np.max(np.linalg.norm(derived_xys - xys, axis=1))
```

```python
# %matplotlib ipympl
# pip install ipympl
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon as PltPolygon

_, ax = plt.subplots(1, 1)

for lid, l in r.map.lanes.items():
    ax.add_patch(PltPolygon(l.polygon.exterior.coords, fc="blue", alpha=0.2, ec="blue"))
ax.plot(*xys.T, label="xys", c="green")
ax.plot(*derived_xys_poly.T, label="derived_xys", c="yellow", linestyle="dashed")
ax.plot(*derived_xys.T, label="derived_xys", c="red", linestyle="dotted")
ax.autoscale()
ax.set_aspect(1)
ax.legend()
ax.set_xlim(-10, 50)
ax.set_ylim(-75, 0)
```
